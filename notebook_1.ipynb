```python
import pandas as pd


trader_data_path = '/content/drive/MyDrive/ds_vaishu/csv_files/historical_data.csv'
sentiment_data_path = '/content/drive/MyDrive/ds_vaishu/csv_files/fear_greed_index.csv'


trader_df = pd.read_csv(trader_data_path)
sentiment_df = pd.read_csv(sentiment_data_path)


print("🔍 Trader Data Preview:")
display(trader_df.head())

print("\n🔍 Sentiment Data Preview:")
display(sentiment_df.head())


print("\n🛠 Trader Data Info:")
trader_df.info()

print("\n🛠 Sentiment Data Info:")
sentiment_df.info()
```

```python
sentiment_df['date'] = pd.to_datetime(sentiment_df['date'])
```

```python
sentiment_df['sentiment'] = sentiment_df['classification'].apply(
    lambda x: 'Fear' if 'Fear' in x else 'Greed'
)
```

```python
sentiment_df = sentiment_df[['date', 'sentiment']]
```

```python
sentiment_df.head()
```

```python
trader_df['Timestamp IST'] = pd.to_datetime(trader_df['Timestamp IST'], format='%d-%m-%Y %H:%M')
```

```python
trader_df['date'] = trader_df['Timestamp IST'].dt.date
```

```python
trader_df['date'] = pd.to_datetime(trader_df['date'])
```

```python
trader_df = trader_df[['Account', 'Side', 'Size USD', 'Closed PnL', 'date']]
```

```python
trader_df.head()
```

```python
merged_df = pd.merge(trader_df, sentiment_df, on='date', how='inner')
```

```python
merged_df.head()
```

```python
summary = merged_df.groupby('sentiment').agg({
    'Size USD': 'sum',         # total volume
    'Account': 'count',        # number of trades
    'Closed PnL': 'mean'       # avg profit/loss per trade
}).rename(columns={
    'Size USD': 'Total Volume (USD)',
    'Account': 'Number of Trades',
    'Closed PnL': 'Avg Closed PnL'
})

display(summary)
```

```python
import matplotlib.pyplot as plt

# Total Volume Bar Chart
summary['Total Volume (USD)'].plot(kind='bar', title='Total Trade Volume by Market Sentiment')
plt.ylabel('Total Volume (USD)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/ds_vaishu/outputs/volume_by_sentiment.png')
plt.show()
```

```python
summary['Number of Trades'].plot(kind='bar', title='Number of Trades by Market Sentiment')
plt.ylabel('Number of Trades')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/ds_vaishu/outputs/num_trades_by_sentiment.png')
plt.show()
```

```python
summary['Avg Closed PnL'].plot(kind='bar', title='Average Profit/Loss by Market Sentiment')
plt.ylabel('Avg Closed PnL per Trade')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/ds_vaishu/outputs/avg_pnl_by_sentiment.png')
plt.show()
```
